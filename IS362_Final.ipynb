Import modules for assignment

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Now we'll import the csv data so that we can read it and convert it to a DataFrame

```python
music = pd.read_csv('https://raw.githubusercontent.com/staceyvo/IS362_Final/master/featuresdf.csv', index_col=False)
spotify = pd.DataFrame(music)
spotify
```

We'll clean the names of the columns up a bit.

```python
spotify.rename(columns=
                {'id' : 'ID',
                 'name' : 'Name', 
                 'artists' : 'Artists',
                 'danceability':'Danceability',
                 'energy': 'Energy',
                 'key':'Key',
                 'loudness':'Loudness',
                 'mode':'Mode',
                 'speechiness':'Speechiness',
                 'acousticness':'Acousticness',
                 'instrumentalness':'Instrumentalness',
                 'liveness':'Liveness',
                 'valence':'Valence',
                 'tempo':'Tempo',
                 'duration_ms':'Duration_ms',
                 'time_signature':'Time Signature'
                }, inplace=True)
spotify.head()
```

We will inspect the data to see if it needs to be cleaned before we can use it.

```python
spotify.dropna()
```

We'll inspect the data further, using the describe function to learn more details about the dataset.

```python
attributes = spotify.describe()
attributes
```

Now that our data has been inspected, let's take a look at the top artists from this spotify dataset.

```python
sns.set_palette('pastel')
plt.subplots(figsize=(12,10))
artist = spotify['Artists'].groupby(spotify['Artists']).count().sort_values(ascending=False)
plt.xlabel('Artists',size=12)
plt.xticks(size=15)
plt.ylabel('Count',size=12)
plt.title('Artists with Most Songs in Top 100',size=20)
artist.head(10).plot('bar')
```

Now we'll chart the mean Dancibility of a song and see what that looks like.

```python
print('Mean value for Danceability:', spotify['Danceability'].mean())
sns.distplot(spotify['Danceability'])
plt.xlabel('Danceability',size=12)
plt.show()
```

Finally, let's combine these all these tables into one graph.  This will show us what makes certain songs more popular than others.

```python
data= spotify.drop(['ID','Name','Artists','Tempo','Duration_ms','Key','Loudness','Time Signature'], axis=1)
```

```python
data.mean().plot.bar()
plt.title('Mean Values of Audio Features')
plt.show()
```

Popular songs all seem to have high energy and danceability, with low speechiness to them.  These characteristics or attributes, have a positive impact on individuals which is why they are played more often.
